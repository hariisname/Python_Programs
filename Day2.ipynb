#### Operators
- Operators are used to perform operations on values
- And this work with variables and data types
- This are mainly used to perform calculations and comparisons
- Widely used in real-time programming

Simple Terms
- Operators are symbols that perform operations between two values or variables.

##### Operatoers Types
- Arithmetic → calculations
- Relational → comparisons
- Assignment → assign values
- Logical → conditions
- Membership → check presence

#### Arithmetic Operators
Arithmetic operators are used to perform mathematical operations on numbers.
- Addition + → adds values
- Subtraction - → subtracts values
- Multiplication * → multiplies values
- Division / → gives decimal result
- Modulus (remainder) % → gives remainder
- Floor division // → removes decimal part
- Exponent (power) ** → power calculation

```python
a = 10
b = 3

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Modulus:", a % b)
print("Floor Division:", a // b)
print("Power:", a ** b)
```

### Relational Operstors
- Relational operators are used to compare two values.
- They return True or False as output.

- Equal to == → checks if values are same

- Not equal to != → checks if values are different

- Greater than > → checks which value is bigger

- Less than < → checks which value is smaller

- Greater than or equal to >= → checks greater or equal

- Less than or equal to <= → checks less or equal

```python
a = 10
b = 20

print(a == b)   # equal to
print(a != b)   # not equal
print(a > b)    # greater than
print(a < b)    # less than
print(a >= b)   # greater than or equal
print(a <= b)   # less than or equal
```

#### Assignment Operators
- Assignment operators are used to assign values to variables.
- They can also update the value using an operation.

- = → assigns value
- += → adds value and stores result
- -= → subtracts and stores
- *= → multiplies and stores
- /= → divides and stores
- %= → stores remainder

```python
a = 10    # initial value is a = 10
print("Initial value:", a)

a += 5   # a = a+5 output is 15 
print("After += :", a)

a -= 2   # a = a - 2 :15 - 2 = 13 
print("After -= :", a)

a *= 3   # a= 13 * 3 = a *= 3 
print("After *= :", a)

a /= 2
print("After /= :", a)

a %= 4
print("After %= :", a)

# this is just another way to assign the values to the variables
```

```python
# Real time example
salary = 30000

salary += 5000   # increment
salary -= 2000   # deduction

print("Final salary =", salary)
```

#### Logical Operators in Python
- Logical operators are used to combine multiple conditions.
- They return True or False as the result.

- and → both conditions must be true
- or → at least one condition must be true
- not → reverses True to False and False to True

```python
a = 10
b = 20

print(a > 5 and b > 10)
print(a > 15 or b > 10)
print(not(a > 5))
```

```python
username = "admin"
password = "1234"

if username == "admin" and password == "1234":
    print("Login successful")
else:
    print("Invalid credentials")
```

#### Membership Operators
- Membership operators are used to check whether a value exists in a collection.
- They work with:
- strings
- lists
- tuples
- sets
- dictionaries

we have 2 operators
- in	    Returns True if value is present
- not in	Returns True if value is not present	

```python
numbers = [10, 20, 30, 40]

print(20 in numbers)
print(50 in numbers)

print(10 not in numbers)
print(60 not in numbers)
```

- Explanation in simple words
- in → checks if value exists
- not in → checks if value does not exist

```python
# String example
text = "python programming"

print("python" in text)
print("java" not in text)
```

```python
fruits = ["apple", "banana", "mango"]

if "apple" in fruits:
    print("Apple is available")
else:
    print("Apple not found")
```

- Membership operators are used to check whether a value is present in a sequence.

# Type Conversion

- Type conversion means changing one data type into another.
- string → int
- int → float
- float → string

Types of Type Conversion
- Implicit Type Conversion
- Explicit Type Conversion (Type Casting)

##### Implicit Type Conversion
- Done automatically by Python
- Converts smaller data type to larger data type
- No data loss

```python
a = 10       # int - this is smaller data type
b = 2.5      # float - this is larger data type

c = a + b

print(c)
print(type(c))

# Python converted int → float automatically.
```

##### Explicit Type Conversion (Type Casting)
- Done manually by programmer
- Uses built-in functions

- Common functions:

- int()
- float()
- str()

```python
# Example 1: String to Integer
a = "10"
b = "20"

sum = int(a) + int(b)
print(sum)
```

```python
# Example 2: Integer to Float
x = 5
y = float(x)

print(y)
```

```python
# Example 3: Number to String
age = 25
text = str(age)

print("Age is " + text)
```

#### Important points
- input() always returns string
- Type conversion is required for calculations
- Wrong conversion causes error
