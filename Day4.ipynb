```python
# Import date class to work with dates and timedelta to add days
from datetime import date, timedelta
# Take start date input from user in dd-mm-yyyy format
d1 = input("Enter start date (dd-mm-yyyy): ")
# Take end date input from user in dd-mm-yyyy format
d2 = input("Enter end date (dd-mm-yyyy): ")
# Split the start date string and convert day, month, year to integers
day1, month1, year1 = map(int, d1.split("-"))
# Split the end date string and convert day, month, year to integers
day2, month2, year2 = map(int, d2.split("-"))
# Create date object for start date
start = date(year1, month1, day1)
# Create date object for end date
end = date(year2, month2, day2)
# Calculate total number of days between start and end dates
total_days = (end - start).days
# Loop from 0 to total_days (inclusive)
for i in range(total_days + 1):
    # Add i days to the start date
    current_date = start + timedelta(days=i)
    # Print the date in dd mm yyyy format
    print(current_date.strftime("%d %m %Y"))
```

```python
# Import the pandas library and assign alias 'pd'
import pandas as pd
# Take start date input from the user in dd-mm-yyyy format
start = input("Enter start date (dd-mm-yyyy): ")
# Take end date input from the user in dd-mm-yyyy format
end = input("Enter end date (dd-mm-yyyy): ")
# Convert start date string to datetime object
# dayfirst=True tells pandas that date format is dd-mm-yyyy
start_date = pd.to_datetime(start, dayfirst=True)
# Convert end date string to datetime object
# dayfirst=True ensures correct date interpretation
end_date = pd.to_datetime(end, dayfirst=True)
# Create a range of dates from start_date to end_date
# freq='D' means daily frequency (one day gap)
dates = pd.date_range(start=start_date, end=end_date, freq='D')
# Loop through each date in the generated date range
for d in dates:
    # Format the date as dd mm yyyy and print it
    print(d.strftime("%d %m %Y"))
```

- and Data types decide what operations can be performed on that data.
- Python is a dynamically typed language, so we don‚Äôt need to declare the data type explicitly.
- The Python interpreter identifies the data type at runtime based on the value assigned.
1. Numeric
    1. Integer
    2. Float
2. Boolean
3. None 
4. Strings
5. Collection
    1. List
    2. Tuple
    3. Dictionary
    4. Set

# Numeric Data type
- Numeric data types are used to store number values in Python.
- They mainly include integers (whole numbers) and floats (decimal numbers).

```python
# Program 1 ‚Äì Integer

a = 10
b = -25
print(a)
print(b)
```

```python
# Program 2 ‚Äì Float

x = 10.5
y = 3.14
print(x)
print(y)
```

```python
# Program 3 ‚Äì Integer and Float together

a = 10
b = 2.5
print(a + b)

# Python automatically converts the integer into a float before performing the operation. implicit type conversion
```

# String
- Strings are used to store text data
- Strings are written inside single or double quotes
- Strings are immutable (cannot be changed)
- Strings support indexing and slicing
- Strings have many built-in functions for manipulation

```python
# 1st point
# Strings are used to store text data
# Strings are used to store names, messages, sentences, or any textual information.
# Example:
name = "Python"
message = "Welcome to Python"
print(name)
print(message)

''' 
Is there any length limit for strings in Python?
No ‚Äî Python does NOT have a fixed length limit for strings.
A string can be as long as your system memory allows.

long_text = "a" * 1000000   # 1 million characters
print(len(long_text))

paragraph = """
Python is a high-level programming language
used for data engineering, automation, and AI.
"""
print(paragraph)    # this is valid
'''




# Practical note (important)
# Very large strings:
#   use more memory
#   can slow down performance
# In real projects, we store huge text in files or databases, not variables
```

```python
# 2 point
# Strings are written inside single or double quotes
# Strings can be enclosed using single quotes (' ') or double quotes (" ").
#Example:

city = 'Bangalore'
language = "Python"
print(city)
print(language)
```

```python
# 3rd point
# Strings are immutable (cannot be changed)
# Once a string is created, its characters cannot be modified.
# Example:

text = "Hello"
# text[0] = "h"   # ‚ùå This will give an error
print(text)


# üëâ To change a string, we must create a new string.
```

```python
# 4th point
# Strings support indexing and slicing
# Each character in a string has an index position starting from 0.
# Example (Indexing):

word = "Python"
print(word[0])   # P
print(word[3])   # h


# Example (Slicing):

print(word[1:4])  # yth
```

# 5th point 
- upper() ‚Äì Converts all characters to uppercase
- lower() ‚Äì Converts all characters to lowercase
- capitalize() ‚Äì Converts the first character to uppercase
- title() ‚Äì Converts the first character of each word to uppercase
- strip() ‚Äì Removes spaces from both left and right sides
- lstrip() ‚Äì Removes spaces from the left side
- rstrip() ‚Äì Removes spaces from the right side
- replace() ‚Äì Replaces a substring with another substring
- split() ‚Äì Splits a string into a list
- join() ‚Äì Joins list elements into a string
- find() ‚Äì Returns the index of the first occurrence of a substring
- count() ‚Äì Counts how many times a substring appears
- startswith() ‚Äì Checks if the string starts with a given value
- endswith() ‚Äì Checks if the string ends with a given value
- isdigit() ‚Äì Checks if the string contains only digits
- isalpha() ‚Äì Checks if the string contains only alphabets
- isalnum() ‚Äì Checks if the string contains letters and numbers
- isspace() ‚Äì Checks if the string contains only spaces
- islower() ‚Äì Checks if all characters are lowercase
- isupper() ‚Äì Checks if all characters are uppercase
- swapcase() ‚Äì Converts uppercase to lowercase and vice versa
- len() ‚Äì Returns the length of the string
- center() ‚Äì Centers the string within a specified width
- zfill() ‚Äì Pads the string with leading zeros
- format() ‚Äì Inserts values into a string in a formatted way 

```python
# strip() ‚Äì Removes extra spaces
# lower() ‚Äì Converts text to lowercase
# replace() ‚Äì Replaces words
# split() ‚Äì Breaks string into list
# len() ‚Äì Counts number of elements
# join() ‚Äì Joins list back to string
```

```python
# 1Ô∏è‚É£ strip() ‚Äì Removes extra spaces

# Removes spaces from the beginning and end of a string.

text = "   Hello Python   "
print(text.strip())
```

```python
# lower() ‚Äì Converts text to lowercase

# Converts all characters in the string to lowercase.

text = "Hello PYTHON"
print(text.lower())
```

```python
# replace() ‚Äì Replaces words

# Replaces a specified word with another word.

text = "I like Java"
print(text.replace("Java", "Python"))
```

```python
# 4Ô∏è‚É£ split() ‚Äì Breaks string into list

# Splits a string into a list of words using spaces by default.

text = "Python is easy"
print(text.split())
```

```python
# 5Ô∏è‚É£ len() ‚Äì Counts number of elements

# Returns the number of elements in a string or list.

text = "Python"
print(len(text))

# Example 2

words = ["Python", "is", "easy"]
print(len(words))
```

```python
# join() ‚Äì Joins list back to string

# Joins list elements into a single string using a separator.

words = ["Python", "is", "easy"]
print(" ".join(words))
print(words)
```
