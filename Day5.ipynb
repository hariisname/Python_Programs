# List
- List can access multiple data types
- List is enclosed with square brackets ([])
- Values are defined using comma-separated values
- List is mutable (we can modify it as per our requirements)
- List allows duplicate values

```python
# List can access multiple data types
my_list = [10, 2.5, "Python", True]  # int, float, String and Boolean
my_list
```

```python
# List is enclosed with square brackets ([])
numbers = [1, 2, 3, 4]
numbers
```

```python
# Values are defined using comma-separated values
fruits = ["apple", "banana", "mango"]
fruits
```

```python
# List is mutable (we can modify it)
nums = [1, 2, 3]
print(nums)
nums[1] = 10
print(nums)
```

```python
# List allows duplicate values
marks = [80, 90, 80, 70]
marks
```

```python
# Accessing Elements from a List
# Accessing elements using index (index starts from 0)
fruits = ["apple", "banana", "mango"]
print(fruits[0])
print(fruits[2])


#Elements are accessed using index positions.
```

```python
# Slicing (extracting a portion of existing data)
nums = [1, 2, 3, 4, 5]
print(nums[1:4])

# Slicing returns a sub-list from the original list.
```

```python
# Reassigning List Elements
# Modifying an element in the list
nums = [10, 20, 30]
nums[1] = 25
print(nums)


# Lists are mutable, so elements can be changed.
```

```python
# Deleting Elements
# Deleting an element using del
nums = [10, 20, 30, 40]
del nums[2]
print(nums)


# Elements can be removed using del.
```

```python
# Multidimensional List
# List inside another list
matrix = [[1, 2], [3, 4]]
print(matrix[0][1])


# A list containing other lists is called a multidimensional list.
```

```python
# Basic Operations on List
# Basic list operations
a = [1, 2]
b = [3, 4]

print(a + b)        # concatenation
print(a * 2)        # repetition
print(len(a))       # length
print(min(b))       # minimum
print(max(b))       # maximum
print(sum(b))       # sum
print(2 in a)       # membership in and not in


# Lists support many built-in operations.
```

```python
# Iteration over List
# Iterating through list elements
nums = [10, 20, 30]
for i in nums:
    print(i)


# Iteration allows accessing elements one by one.
```

```python
# List Comprehension : it is another way of creating list 
# Creating a list using list comprehension    
squares = [x*x for x in range(1, 6)]
print(squares)

numbers = [i for i in range(1,6)]
print(numbers)

# List comprehension is a short and clean way to create lists.
```

##### List Built-in List Methods

```python

# append()
# Adds element at the end
nums = [1, 2]
nums.append(3)
print(nums)
```

```python
# extend()
# Adds multiple elements
nums = [1, 2]
nums.extend([3, 4])
print(nums)
```

```python
# insert()
# Inserts element at a specific position
nums = [1, 3]
nums.insert(1, 2)
print(nums)
```

```python
# remove()
# Removes first occurrence of value
nums = [1, 2, 3, 2]
nums.remove(2)
print(nums)
```

```python
# pop()
# Removes and returns element
nums = [1, 2, 3]
nums.pop()
print(nums)
```

```python
# sort()
# Sorts list in ascending order
nums = [3, 1, 2]
nums.sort()
print(nums)
```

```python
# count()
# Counts occurrences of an element
nums = [1, 2, 2, 3]
print(nums.count(2))
```

```python
# index()
# Returns index of element
nums = [10, 20, 30]
print(nums.index(20))
```

```python
# reverse()
# Reverses list order
nums = [1, 2, 3]
nums.reverse()
print(nums)
```

## Tuple

```python
# Tuple is similar to List data structure
# Tuple stores multiple values like a list
t = (10, 20, 30)
t
```

```python
# Tuple is immutable (cannot change, add, or reassign elements)
# Trying to modify a tuple element
t = (1, 2, 3)
# t[0] = 10   # This will cause an error

t
# Once created, tuple elements cannot be modified.
```

```python
# Tuple is defined using parentheses ()
colors = ("red", "green", "blue")

colors
```

```python
# Tuple can store multiple data types
data = (10, 2.5, "Python", True)

data
```

```python
# Tuple structure cannot be modified directly

t = (1, 2, 3)
lst = list(t)
lst.append(4)
t = tuple(lst)
print(t)


# To modify a tuple, convert it to a list, make changes, then convert it back.
```

```python
# Accessing Elements from Tuple
# Accessing tuple elements using index
t = (10, 20, 30)
print(t[0])


# Elements are accessed using index positions (starting from 0).
```

```python
# Indexing and Slicing
# Indexing : accesses one element
t = (1, 2, 3, 4, 5)
print(t[2])

# Slicing : extracts a portion of the tuple.
print(t[1:4])
```

```python
# Reassigning Tuple Elements
t = (1, 2, 3)
# t[0] = 10   # Error


# Tuple elements cannot be reassigned because tuples are immutable.
```

```python
t=(1,2,3)
t1 = list(t)
t1[1] = 20
t=tuple(t1) 
t
```

```python
# Deleting Elements
t = (1, 2, 3)
# del t[1]    # Error


# You cannot delete individual elements from a tuple.
# Only the entire tuple can be deleted:
```

```python
# Nested Tuples
# Tuple inside another tuple
t = (1, (2, 3), 4)
print(t[1][0])


# A tuple containing another tuple is called a nested tuple.
```

```python
# Basic Operations on Tuple
a = (1, 2)
b = (3, 4)

print(a + b)        # concatenation
print(a * 2)        # repetition
print(len(a))       # length
print(min(b))       # minimum
print(max(b))       # maximum
print(sum(b))       # sum
print(2 in a)       # membership
```

```python
# Not Required
# Iteration
for i in a:
    print(i)


# Tuples support common operations like lists.
```

```python
# Built-In Methods
# count()
t = (1, 2, 2, 3)
print(t.count(2))

# Returns how many times an element appears.
```

```python
# index()
t = (10, 20, 30)
print(t.index(20))


#  Returns the index of the first occurrence.
```

```python
# enumerate()
t = ("a", "b", "c")

for index, value in enumerate(t):
    print(index, value)


# Returns index and value together while iterating.
```

```python
# Why does Tuple have very few built-in methods?
#Simple explanation:
# - Tuples are immutable, so Python does not provide methods that modify data.
#Key reasons:

# - No adding elements
# - No removing elements
# - No updating elements

# Only read-only operations are allowed

# Thatâ€™s why tuples have only count() and index() as built-in methods.
```

```python
# Main reason using tuples - 
# Tuples are faster than lists in terms of performance
# Tuples are immutable; once created, they cannot be modified
# Tuples are more memory-efficient than lists
```
